```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Loading dataset
df=pd.read_csv(r"C:\Users\LENOVO\Downloads\SampleSuperstore.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
# Data Analysis
sns.set_style("whitegrid")
sns.pairplot(df,hue="Region",palette='bwr')
```

```python
df['Ship Mode'].unique()
```

```python
segment=df.Segment.value_counts().reset_index()
segment.columns=("Segment","Count")
segment
```

```python
# Pie Charts
plt.pie(x="Count",labels="Segment",data=segment,radius=3,autopct="%.2f",pctdistance=0.4)
```

```python
plt.figure(figsize=(9,5))
plt.title('CATEGORIES VS REGION')
sns.countplot(x=df['Category'],hue=df['Region'],palette='rocket')
plt.xticks()
```

```python
# Finding columns with high sales
high_sale=df[df["Sales"]>3000]
high_sale.head()
```

```python
# Finding states with high sales
plt.figure(figsize=(10,6))
sns.barplot(x=high_sale["Sales"],y=high_sale["State"],data=df)
```

```python
# Finding maximum value of sale
df.Sales.max()
```

```python
# Finding minimum value of sale
df.Sales.min()
```

```python
# Display states with the lowest sales data
plt.figure(figsize=(9,5))
low_sale=df[df["Sales"]<1]
sns.barplot(y=low_sale["State"],x=low_sale["Sales"],data=low_sale,palette='rocket')
```

```python
# Data for low profit
low_profit=df[df["Profit"]<1]
low_profit.head()
```

```python
# Heatmap correlation matrix
data=['Sales','Quantity','Discount','Profit']
plt.figure(figsize=(9,5))
sns.heatmap(df[data].corr(),annot=True )
```

### Conclusion:
It is clear from the heatmap correlation visualization that sales and profit are high positively correlated, profit and discount are negatively correlated
